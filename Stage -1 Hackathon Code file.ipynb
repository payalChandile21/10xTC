```python
import pandas as pd

# Replace 'your_dataset.csv' with the actual file path
avengers_data = 'Avengers_Assemble.csv'

# Read CSV file into a DataFrame
df = pd.read_csv(avengers_data)

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df['Gender'] = df['Gender'].str.lower().str.strip()
```

```python
gender_count=df['Gender'].value_counts()
print(gender_count)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import pyttsx3

# Initialize the TTS engine
engine = pyttsx3.init()

# Set the properties (optional)
engine.setProperty('rate', 150)  # Speed of speech
engine.setProperty('volume', 0.9)  # Volume level (0.0 to 1.0)

# Variable containing text to be converted to speech
#custom_message = "I am a variable containing custom text."

# Convert text to speech
#engine.say(custom_message)

# Wait for the speech to finish
#engine.runAndWait()

# Function to list all Avengers with a specific ability
def list_avengers_by_ability():
    ability = input("Enter the ability to search for: ")
    engine.say(f"Here are the Avengers with the ability {ability}")
    engine.runAndWait()
    avengers_with_ability = df[df['ability'] == ability]
    print(f"Avengers with the ability '{ability}':")
    print(avengers_with_ability[['name', 'ability', 'work_affiliation']])

# Function to list all Avengers older than a certain age
def list_avengers_older_than_age():
    age_limit = int(input("Enter the age limit: "))
    engine.say(f"Here are the Avengers with the age greater than {age_limit}")
    engine.runAndWait()
    older_avengers = df[df['age'] > age_limit]
    print(f"Avengers older than {age_limit} years:")
    print(older_avengers[['name', 'age', 'work_affiliation']])

# Function to display all Avengers sorted by age
def display_avengers_sorted_by_age():
    sorted_by_age = df.sort_values(by='age', ascending=True)
    engine.say("Here is the list of Avengers by Age ")
    engine.runAndWait()
    print("All Avengers sorted by age:")
    print(sorted_by_age[['name', 'ability', 'age', 'work_affiliation']])

# Function to list all Avengers with a specific work affiliation
def list_avengers_by_work_affiliation():
    affiliation = input("Enter the work affiliation to search for: ")
    engine.say(f"Here are the Avengers with the Work Affiliation {affiliation}")
    engine.runAndWait()
    avengers_with_affiliation = df[df['work_affiliation'] == affiliation]
    print(f"Avengers with work affiliation '{affiliation}':")
    print(avengers_with_affiliation[['name', 'ability', 'age', 'enemy']])

# Function to list all Avengers with a specific enemy
def list_avengers_by_enemy():
    enemy = input("Enter the enemy to search for: ")
    engine.say(f"Here are the Avengers whose enemy is {enemy}")
    engine.runAndWait()
    avengers_with_enemy = df[df['enemy'] == enemy]
    print(f"Avengers with enemy '{enemy}':")
    print(avengers_with_enemy[['name', 'enemy']])

# Function to display a pie chart for the distribution of Avengers' abilities
def plot_ability_pie_chart():
    engine.say("Here is the Pie Chart for the distribution of Avengers' abilities")
    engine.runAndWait()
    plt.figure(figsize=(8, 8))
    ability_counts = df['ability'].value_counts()
    plt.pie(ability_counts, labels=ability_counts.index, autopct='%1.1f%%', startangle=90)
    plt.title("Distribution of Abilities Among Avengers")
    plt.show()

# Function to display a bar graph for the distribution of Avengers based on their enemies
def plot_enemy_bar_graph():
    engine.say("Here is the bar graph for the distribution of Avengers based on their enemies")
    engine.runAndWait()
    plt.figure(figsize=(12, 6))
    enemy_counts = df['enemy'].value_counts()
    enemy_counts.plot(kind='bar', color='skyblue')
    plt.xlabel("Enemy")
    plt.ylabel("Number of Avengers")
    plt.title("Number of Avengers Associated with Each Enemy")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()

# Function to display a pie chart for the distribution of Avengers based on their work affiliation
def plot_affiliation_pie_chart():
    engine.say("Here is the Pie Chart for the distribution of Avengers based on their Work Affliliation")
    engine.runAndWait()
    plt.figure(figsize=(8, 8))
    affiliation_counts = df['work_affiliation'].value_counts()
    plt.pie(affiliation_counts, labels=affiliation_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral', 'lightgreen'])
    plt.title("Distribution of Avengers Based on Work Affiliation")
    plt.show()

# Function to display a bar chart for the distribution of Avengers based on their availability
def plot_availability_bar_chart():
    engine.say("Here is the bar graph for the distribution of Avengers based on their availability")
    engine.runAndWait()
    plt.figure(figsize=(8, 6))
    availability_counts = df['availability'].value_counts()
    availability_counts.plot(kind='bar', color=['lightcoral', 'lightgreen'])
    plt.xlabel("Availability")
    plt.ylabel("Number of Avengers")
    plt.title("Distribution of Avengers Based on Availability")
    plt.xticks(rotation=0)
    plt.show()
    
    #Bonus Question
    #Function to Count Male and Female

def Avengers_Gender_count():
    engine.say("Here is the total number of Male and Female Avengers")
    engine.runAndWait()
    
    gender_count = df['Gender'].value_counts()
    print(gender_count)
    
    # Plotting the bar chart
    plt.figure(figsize=(8, 6))
    plt.bar(gender_count.index, gender_count.values)
    plt.title('Gender Distribution of Avengers')
    plt.xlabel('Gender')
    plt.ylabel('Count')
    plt.show()

# Menu-driven program
while True:
    print("\nMenu:")
    print("1. List Avengers with a specific ability")
    print("2. List Avengers older than a certain age")
    print("3. Display all Avengers sorted by age")
    print("4. List Avengers with a specific work affiliation")
    print("5. List Avengers with a specific enemy")
    print("6. Display a pie chart for the distribution of Avengers' abilities")
    print("7. Display a bar graph for the distribution of Avengers based on their enemies")
    print("8. Display a pie chart for the distribution of Avengers based on their work affiliation")
    print("9. Display a bar chart for the distribution of Avengers based on their availability")
    print("10.Display the total number of Male and Female Avengers")
    print("0. Exit")

    choice = input("Enter your choice (0-9): ")

    if choice == '1':
        list_avengers_by_ability()
    elif choice == '2':
        list_avengers_older_than_age()
    elif choice == '3':
        display_avengers_sorted_by_age()
    elif choice == '4':
        list_avengers_by_work_affiliation()
    elif choice == '5':
        list_avengers_by_enemy()
    elif choice == '6':
        plot_ability_pie_chart()
    elif choice == '7':
        plot_enemy_bar_graph()
    elif choice == '8':
        plot_affiliation_pie_chart()
    elif choice == '9':
        plot_availability_bar_chart()
    elif choice == '10':
        Avengers_Gender_count()        
    elif choice == '0':
        engine.say("Goodbye! Have a great day!")
        engine.runAndWait()
        break
    else:
        print("Invalid choice. Please enter a number from 0 to 9.")

```
